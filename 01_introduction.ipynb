```python
import json
```

```python
# Lets write a function to load the data
def load_data(filename):
    with open(filename,'r')as f:
        data = json.load(f)
        return data
```

```python
data = load_data("data.json")
data
```

```python
# write a fuction to display user and their connections
def display_users(data):
    print("users and their connections:")
    for user in data['users']:
        print(f"ID:{user['id']} - {user['name']} is friends with : {user['friends']}  and liked_pagers are {user['liked_pages']}")
    print("\npages Information:")
    for page in data['pages']:
        print(f"ID:-{page['id']}: {page['name']}")

display_users(data)        
```
